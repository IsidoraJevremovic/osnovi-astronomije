```python
from ephem import *
from pandas import *
import matplotlib.pyplot as plt
```

```python
posmatrac = Observer()
posmatrac.lat = '44.8125'
posmatrac.lon = '20.4612'

sunce = Sun(posmatrac)

sunce_azimuti = []
sunce_visine = []

merkur_azimuti = []
merkur_visine = []

jupiter_azimuti = []
jupiter_visine = []

pluton_azimuti = []
pluton_visine = []

venera_azimuti = []
venera_visine = []

uran_azimuti = []
uran_visine = []

mars_azimuti = []
mars_visine = []

saturn_azimuti = []
saturn_visine = []

neptun_azimuti = []
neptun_visine = []

for datum in date_range('2024/4/26 9:00:00', '2025/4/26 9:00:00'):
  posmatrac.date = datum
  sunce = Sun(posmatrac)
  merkur = Mercury(posmatrac)
  venera = Venus(posmatrac)
  mars = Mars(posmatrac)
  jupiter = Jupiter(posmatrac)
  saturn = Saturn(posmatrac)
  uran = Uranus(posmatrac)
  neptun = Neptune(posmatrac)
  pluton = Pluto(posmatrac)

  sunce_azimuti.append(sunce.az * 180 / 3.1415)
  sunce_visine.append(sunce.alt * 180 / 3.1415)

  merkur_azimuti.append(merkur.az * 180 / 3.1415)
  merkur_visine.append(merkur.alt * 180 / 3.1415)

  venera_azimuti.append(venera.az * 180 / 3.1415)
  venera_visine.append(venera.alt * 180 / 3.1415)

  mars_azimuti.append(mars.az * 180 / 3.1415)
  mars_visine.append(mars.alt * 180 / 3.1415)

  jupiter_azimuti.append(jupiter.az * 180 / 3.1415)
  jupiter_visine.append(jupiter.alt * 180 / 3.1415)

  saturn_azimuti.append(saturn.az * 180 / 3.1415)
  saturn_visine.append(saturn.alt * 180 / 3.1415)

  uran_azimuti.append(uran.az * 180 / 3.1415)
  uran_visine.append(uran.alt * 180 / 3.1415)

  neptun_azimuti.append(neptun.az * 180 / 3.1415)
  neptun_visine.append(neptun.alt * 180 / 3.1415)

  pluton_azimuti.append(pluton.az * 180 / 3.1415)
  pluton_visine.append(pluton.alt * 180 / 3.1415)
```

```python
plt.figure()
plt.title('Kretanje unutrasnjih nebeskih tela u periodu od 26.4.2024. do 26.4.2025. svakog dana u 9h po grinickom vremenu')
plt.xlabel('Azimut (u stepenima)')
plt.ylabel('Visina (u stepenima)')
plt.scatter(sunce_azimuti, sunce_visine, color = 'blue', marker = '.', label = 'Sunce')
plt.scatter(merkur_azimuti, merkur_visine, color = 'green', marker = '.', label = 'Merkur')
plt.scatter(venera_azimuti, venera_visine, color = 'orange', marker = '.', label = 'Venera')
plt.scatter(mars_azimuti, mars_visine, color = 'red', marker = '.', label = 'Mars')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.title('Kretanje spoljasnjih nebeskih tela u periodu od 26.4.2024. do 26.4.2025. svakog dana u 9h po grinickom vremenu')
plt.xlabel('Azimut (u stepenima)')
plt.ylabel('Visina (u stepenima)')
plt.scatter(jupiter_azimuti, jupiter_visine, color = 'pink', marker = '.', label = 'Jupiter')
plt.scatter(saturn_azimuti, saturn_visine, color = 'purple', marker = '.', label = 'Saturn')
plt.scatter(uran_azimuti, uran_visine, color = 'violet', marker = '.', label = 'Uran')
plt.scatter(neptun_azimuti, neptun_visine, color = 'navy', marker = '.', label = 'Neptun')
plt.scatter(pluton_azimuti, pluton_visine, color = 'cyan', marker = '.', label = 'Pluton')
plt.legend()
plt.show()
```
