```python
from ephem import *
from pandas import *
import matplotlib.pyplot as plt
```

```python
posmatrac = Observer()
posmatrac.lat = '44.8125'
posmatrac.lon = '20.4612'

sunce = Sun(posmatrac)

azimuti = []
visine = []

prolecna_ravnodnevnica = []
jesenja_ravnodnevnica = []
zimska_kratkodnevnica = []
letnja_dugodnevnica = []

for datum in date_range('2024/1/1 12:00:00', '2024/12/31 12:00:00'):
  posmatrac.date = datum
  sunce = Sun(posmatrac)

  azimuti.append(sunce.az * 180 / pi)
  visine.append(sunce.alt * 180 / pi)

  if datum == '2024/9/22':
    jesenja_ravnodnevnica.append(sunce.az * 180 / pi)
    jesenja_ravnodnevnica.append(sunce.alt * 180 / pi)
  if datum == '2024/3/20':
    prolecna_ravnodnevnica.append(sunce.az * 180 / pi)
    prolecna_ravnodnevnica.append(sunce.alt * 180 / pi)
```

```python
plt.figure()
plt.title('Analema za posmatraca u Beogradu u podne')
plt.plot(azimuti, visine, color = 'orange')
plt.show()
```
