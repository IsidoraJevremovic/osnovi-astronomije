```python
from ephem import *
from pandas import *
import matplotlib.pyplot as plt
```

```python
posmatrac = Observer()
posmatrac.lat = '44.8125'
posmatrac.lon = '20.4612'

sunce = Sun()

datumi = []
odstupanja = []
nula = []

for datum in date_range('2024/1/1', '2024/12/31'):
  posmatrac.date = datum
  datumi.append(datum)
  nula.append(0)

  if datum in date_range('2024/3/31', '2024/10/27'):
    odstupanja.append((posmatrac.next_transit(sunce).tuple()[3] - 11  + 1) * 60 + posmatrac.next_transit(sunce).tuple()[4] + posmatrac.next_transit(sunce).tuple()[5] / 60)
  else:
    odstupanja.append((posmatrac.next_transit(sunce).tuple()[3] - 11 ) * 60 + posmatrac.next_transit(sunce).tuple()[4] + posmatrac.next_transit(sunce).tuple()[5] / 60)
```

```python
plt.figure()
plt.title('Odstupanje gornje kulminacije pravog Sunca po lokalnom vremenu u Beogradu 2024.')
plt.plot(datumi, odstupanja, color = 'y')
plt.plot(datumi, nula, color = "g")
plt.show()
```
