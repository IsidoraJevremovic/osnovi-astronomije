```python
from ephem import *
from pandas import *
import matplotlib.pyplot as plt
import pytz
from datetime import datetime

posmatrac = Observer()
posmatrac.lat = '44.8125'
posmatrac.lon = '20.4612'

sunce = Sun()

datumi = []
odstupanja_u_minutima = []
lokalno_vreme = []



for datum in date_range('2024-01-01', '2024-12-31'):
    pravo_podne = posmatrac.previous_transit(sunce, datum)
    podne = posmatrac.next_transit(sunce, datum)
    odstupanje = (podne - pravo_podne)*24*60
    datumi.append(datum)
    odstupanja_u_minutima.append(odstupanje)


plt.plot(datumi, odstupanja_u_minutima)
#plt.plot(datumi, len(datumi)*[0])
plt.title('Odstupanje gornje kulminacije pravog Sunca po lokalnom vremenu (2024)')
plt.xlabel('Datum')
plt.ylabel('Odstupanje (minuti)')
plt.show()

```
