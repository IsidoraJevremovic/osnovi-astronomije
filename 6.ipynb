```python
from ephem import *
from pandas import *
import matplotlib.pyplot as plt
```

```python
posmatrac = Observer()
posmatrac.lat = '44.8125'
posmatrac.lon = '20.4612'

sunce = Sun()

datumi = []
zalasci = []

for datum in date_range('2024/1/1', '2024/12/31'):
  posmatrac.date = datum
  datumi.append(datum)

  if datum in date_range('2024/3/31', '2024/10/27'):
    zalasci.append((posmatrac.next_setting(sunce).tuple()[3] + 1 + 1) + posmatrac.next_setting(sunce).tuple()[4] / 60 + posmatrac.next_setting(sunce).tuple()[5] / 3600)
  else:
    zalasci.append((posmatrac.next_setting(sunce).tuple()[3] + 1) + posmatrac.next_setting(sunce).tuple()[4] / 60 + posmatrac.next_setting(sunce).tuple()[5] / 3600)
```

```python
plt.figure()
plt.title('Zalasci Sunca po lokalnom vremenu u Beogradu 2024.')
plt.plot(datumi, zalasci, color = 'g')
plt.show()
```
